```python
import pandas as pd
import numpy as np
import math
from numpy import dot
from numpy.linalg import norm
```

```python
data = pd.read_csv(r'C:\Users\moon_\DataAnalytics/data_week5.txt')
data
```

data에서 user별 점수의 분포를 나타내는 matrix를 생성한다.

```python
df1 = data.pivot_table('Score', index = 'User', columns='Item')
df1
```

User X에 대한 정보를 데이터프레임화하여, df1과 합병한다.

```python
user_x_df =pd.DataFrame({2: [2], 3: [1], 5: [3], 8: [4], 9: [3]})

user_x_df
```

```python
df = pd.concat([df1, user_x_df])
df
```

df에 '평균값'이라는 column을 새로 생성하여 각 행마다 평균을 계산한다. (이때 Nan은 없는 값으로 간주하고, 표본의 갯수에 포함하지 않는다.)

```python
for i in range(len(df)):
    print(df.iloc[i, :-1].mean())
```

```python
df['평균값'] = df.iloc[:, 0:10].mean(axis=1)
df
```

이제 1열부터 10열까지 (item 1~10)의 rating 값에서 평균값을 제거하여 tendency를 제거한다.

```python
for i in range(len(df)):
    for j in range(10):
        if math.isnan(df.iloc[i, j]) == True:
            pass
        else:
            df.iloc[i, j] = df.iloc[i, j] - df.iloc[i, 10]
```

애초에 i행 j열에 값이 null인 것들은 계산을 할 필요가 없다. null이 아닌 값에 평균값을 빼주어 값을 산출한다.

```python
df = df.iloc[:, :10]
df
```

```python
df23589 = df[[2, 3 ,5 , 8 , 9]]
df23589
```

코사인 유사도를 산출해야하는데, User X는 2, 3, 5, 8, 9에 관해서만 ratio를 매겼으므로, 2, 3, 5, 8, 9 item 값의 df를 산출한다.

```python
df23589 = df23589.replace(np.nan, 0)
df23589
```

Nan값은 유사도를 산출할 수 없으므로, 0으로 대체한다.

```python
user_x = []
for i in range(10):
    user_x.append(df23589.iloc[-1, :])

user_x = np.array(user_x)
user_x
```

```python
for i in range(10):
    for j in range(len(user_x[0])):
        if (df23589.iloc[i, j]) == 0:
            user_x[i][j] = 0
```

```python
user_x
```

df23589에서 0값을 가지는 i행 j열의 value들에 맞추어 0으로 변환시켜준다.

```python
#코사인 유사도를 구하는 함수
def cos_sim(a, b):
    return dot(a, b) / (norm(a) * norm(b))
```

```python
user1_10 = df23589.iloc[:10, :]
user1_10
```

```python
for i in range(len(df)-1):
    print('User X와 User', i+1, '의 코사인 유사도: ', cos_sim(user1_10.iloc[i, :], user_x[i]))
```
